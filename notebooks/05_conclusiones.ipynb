# 1. Introducción

En esta última etapa del Proyecto Integrador se presentan las conclusiones obtenidas a partir del proceso completo de minería de datos desarrollado sobre el dataset seleccionado.

Las conclusiones sintetizan la información obtenida durante las etapas de inspección, limpieza, análisis exploratorio y reducción de dimensionalidad mediante PCA, procurando responder a los objetivos planteados al inicio del proyecto.

Asimismo, se identifican las principales limitaciones encontradas durante el desarrollo del trabajo y se proponen posibles líneas de mejora para futuras investigaciones o ampliaciones del análisis.

Es importante destacar que las conclusiones aquí presentadas distinguen claramente entre la evidencia observada durante el análisis, su interpretación y las inferencias finales derivadas del proceso.

# 2. Síntesis del proceso de minería de datos

El desarrollo del proyecto se estructuró en cuatro etapas principales:

- Inspección inicial del dataset para comprender su estructura, variables y calidad de los datos.
- Limpieza y preparación de la información, corrigiendo inconsistencias, estandarizando variables y tratando valores faltantes.
- Análisis exploratorio de datos (EDA), donde se estudiaron distribuciones, relaciones entre variables y patrones relevantes.
- Escalamiento de variables numéricas y aplicación de Análisis de Componentes Principales (PCA), con el objetivo de reducir la dimensionalidad y analizar la estructura interna del conjunto de datos.

Cada una de estas etapas permitió mejorar progresivamente la calidad de la información disponible y facilitar la obtención de conocimiento útil para responder a los objetivos del proyecto.

# 3. Respuesta a los objetivos del proyecto

A partir del proceso completo de análisis puede concluirse que los objetivos planteados fueron alcanzados.

En primer lugar, fue posible comprender la estructura del dataset y mejorar significativamente su calidad mediante un proceso sistemático de limpieza y preparación de datos.

Posteriormente, el análisis exploratorio permitió identificar patrones relevantes, distribuciones de las variables y relaciones entre distintos atributos del conjunto de datos.

Finalmente, la aplicación de técnicas de escalamiento y reducción de dimensionalidad permitió representar la información de manera más compacta, preservando gran parte de la variabilidad original y facilitando la interpretación de la estructura del dataset.

# 4. Principales hallazgos

Entre los principales resultados obtenidos durante el análisis pueden destacarse:

- El proceso de limpieza permitió construir un dataset consistente y apto para posteriores análisis.
- Las variables analizadas presentaron distintos comportamientos estadísticos, evidenciando tanto distribuciones homogéneas como valores extremos en determinados casos.
- El análisis exploratorio permitió detectar relaciones entre variables y comprender mejor la composición del conjunto de datos.
- La reducción de dimensionalidad mediante PCA demostró que gran parte de la información puede representarse utilizando un número reducido de componentes principales.
- La visualización de los componentes principales permitió observar la estructura general del dataset y facilitar futuras tareas de modelado.

# 5. Conclusiones generales

El proceso de minería de datos permitió transformar un conjunto de datos originalmente heterogéneo en una fuente de información organizada, consistente y preparada para posteriores análisis.

La combinación de técnicas de preparación, análisis exploratorio y reducción de dimensionalidad facilitó la identificación de patrones relevantes y proporcionó una comprensión más profunda de la información contenida en el dataset.

Los resultados obtenidos demuestran la importancia de realizar una adecuada preparación de los datos antes de aplicar técnicas analíticas más avanzadas, ya que la calidad de los datos condiciona directamente la calidad de las conclusiones alcanzadas.

En conjunto, el proyecto permitió cumplir los objetivos propuestos y establecer una base sólida para futuras aplicaciones de aprendizaje automático o análisis predictivo.

# 6. Limitaciones

El alcance de las conclusiones se encuentra condicionado por la información disponible en el dataset y por las decisiones metodológicas adoptadas durante el proceso de preparación y análisis de los datos.

Asimismo, algunas variables presentaron limitaciones en cuanto a su nivel de detalle o disponibilidad, lo que restringe el alcance de determinadas interpretaciones.

Finalmente, el análisis desarrollado se centró en técnicas descriptivas y de reducción de dimensionalidad, por lo que no se abordaron tareas predictivas ni modelos supervisados.

# 7. Mejoras futuras

Como posibles líneas de trabajo futuro se propone:

- Incorporar nuevas variables que permitan enriquecer el análisis.
- Actualizar periódicamente el dataset para evaluar la evolución temporal de la información.
- Aplicar técnicas de aprendizaje automático supervisado para desarrollar modelos predictivos.
- Comparar distintos métodos de reducción de dimensionalidad y agrupamiento.
